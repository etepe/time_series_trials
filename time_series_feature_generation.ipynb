```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

 

usdtry_df=pd.read_csv('C:/Users/FON04520/Documents/Python Scripts/data/USDTRY.csv',  header=0, sep=';')
usdtry_df.set_index(pd.to_datetime(usdtry_df['Dates'], infer_datetime_format=True), inplace=True)

 

market_df=pd.read_excel('C:/Users/FON04520/Documents/market_df.xlsx')
market_df.set_index(pd.to_datetime(market_df['Dates'], infer_datetime_format=True),inplace=True)
market_df.drop('Dates', axis=1,inplace=True)

 

#usdtry_df.loc[min(usdtry_df.loc['2018-06'].index)]

 

usdtry_daily=market_df[['USDTRY Curncy']]
usdtry_daily['year']=usdtry_daily.index.year
usdtry_daily['quarter']=usdtry_daily.index.quarter
usdtry_daily['month']=usdtry_daily.index.month
usdtry_daily['week']=usdtry_daily.index.week
usdtry_daily['day']=usdtry_daily.index.day

 

 
usdtry_daily['year_open']=0.0
usdtry_daily['quarter_open']=0.0
usdtry_daily['month_open']=0.0
usdtry_daily['week_open']=0.0

 

a=len(usdtry_daily['USDTRY Curncy'])-1

 

for i in range(0,a):

    #usdtry_daily['week open'][i]

    year_f=usdtry_daily['year']==usdtry_daily['year'][i]
    quarter_f=usdtry_daily['quarter']==usdtry_daily['quarter'][i]
    month_f=usdtry_daily['month']==usdtry_daily['month'][i]
    week_f=usdtry_daily['week']==usdtry_daily['week'][i]

    mask_year=year_f
    mask_quareter=np.logical_and(year_f,quarter_f)
    mask_month=np.logical_and(year_f,month_f)
    mask_week=np.logical_and(year_f,week_f) 

    year_index=usdtry_daily[mask_year].index.min()
    quarter_index=usdtry_daily[mask_quareter].index.min()
    month_index=usdtry_daily[mask_month].index.min()
    week_index=usdtry_daily[mask_week].index.min() 

    year_start=usdtry_daily.loc[year_index][0]
    quarter_start=usdtry_daily.loc[quarter_index][0]
    month_start=usdtry_daily.loc[month_index][0]
    week_start=usdtry_daily.loc[week_index][0]
   

    usdtry_daily['year_open'][i]=year_start
    usdtry_daily['quarter_open'][i]=quarter_start
    usdtry_daily['month_open'][i]=month_start
    usdtry_daily['week_open'][i]=week_start

   

usdtry_daily['YTD_chg']=(usdtry_daily['USDTRY Curncy']/usdtry_daily['year_open'])-1
usdtry_daily['QTD_chg']=(usdtry_daily['USDTRY Curncy']/usdtry_daily['quarter_open'])-1
usdtry_daily['MTD_chg']=(usdtry_daily['USDTRY Curncy']/usdtry_daily['month_open'])-1
usdtry_daily['WTD_chg']=(usdtry_daily['USDTRY Curncy']/usdtry_daily['week_open'])-1
```
